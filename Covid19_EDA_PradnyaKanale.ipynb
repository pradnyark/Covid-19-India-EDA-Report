Importing the libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
%matplotlib inline
import io
```

Reading and printing the dataset

```python
df = pd.read_csv(io.BytesIO(uploaded['Latest Covid-19 India Status.csv']))
print(df)
```

Since all the data is clean, we can visually analyze the data using the plot library and seaborn library 

**Correlation** and **Statewise analysis**

```python
#correlation using heatmap visuals
f,ax = plt.subplots(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, ax=ax)
plt.show()
```

```python
#Top 5 highestCases
highestCases = df.sort_values(by ='Total Cases', ascending = False).head(5)
highestCases
```

```python
#Top 5 highestRecoveryRate
highestRecoveryRate = df.sort_values(by ='Discharged', ascending = False).head(5)
highestRecoveryRate
```

```python
#comparison of the active and death ratio
px.scatter(df, x='Active Ratio',y='Death Ratio', color=df['State/UTs'])
```

```python
plt.subplot(1,3,1)
plt.plot(np.log(df['Total Cases']), np.log(df['Active']), 'go')
plt.title("Active VS Total Cases ",fontdict={'size':12,'color':'green'})
plt.xlabel("Total Cases")
plt.ylabel("Active")

plt.subplot(1,3,2)
plt.plot(np.log(df['Total Cases']), np.log(df['Discharged']), 'b*')
plt.title("Discharged VS Total Cases",fontdict={'size':12,'color':'blue'})
plt.xlabel("Total Cases")
plt.ylabel("Discharged")

plt.subplot(1,3,3)
plt.plot(np.log(df['Total Cases']), np.log(df['Deaths']), 'go')
plt.title("Deaths VS Total Cases",fontdict={'size':12,'color':'red'})
plt.xlabel("Total Cases")
plt.ylabel("Deaths")

plt.tight_layout()
plt.show
```
